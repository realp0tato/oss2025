```python
# prompt: /content/부산도시공사_차량정보.csv 이 데이터에서 '누적주행거리값'의 전체를 그래프로 나타내서 보여주고 최대값, 최소값에 값을 표시해줘 그리고 구분배기량도 그래프로나타내서 보여주고 최대값, 최소값에 값을 표시해서 나타내줘

import os
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files  # Import files for uploading

file_path = '/content/부산도시공사_차량정보.csv'

# Check if the file exists
if not os.path.exists(file_path):
    print(f"File not found at {file_path}. Please upload the file.")
    uploaded = files.upload()  # Upload the file
    # Assuming the uploaded file has the same name
    file_path = next(iter(uploaded))

# CSV 파일 읽기 (encoding='cp949' 추가)
try:
    df = pd.read_csv(file_path, encoding='cp949')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='utf-8')  # 또는 다른 인코딩 시도

# (Rest of your code remains the same)
# '누적주행거리값' 컬럼으로 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(df['누적주행거리값'])
plt.xlabel('Index')
plt.ylabel('Cumulative Driving Distance Value')
plt.title('Cumulative Driving Distance Value Graph')

# 최대값, 최소값 표시
max_value = df['누적주행거리값'].max()
min_value = df['누적주행거리값'].min()

plt.annotate(f'Max: {max_value}', xy=(df['누적주행거리값'].idxmax(), max_value), xytext=(df['누적주행거리값'].idxmax(), max_value + 1000), arrowprops=dict(arrowstyle='->'))
plt.annotate(f'Min: {min_value}', xy=(df['누적주행거리값'].idxmin(), min_value), xytext=(df['누적주행거리값'].idxmin(), min_value - 1000), arrowprops=dict(arrowstyle='->'))

plt.show()


# '구분배기량' 컬럼으로 그래프 생성 (만약 '구분배기량' 컬럼이 숫자형이 아니라면 변환 필요)
if pd.api.types.is_numeric_dtype(df['구분배기량']):
    plt.figure(figsize=(10, 6))
    plt.plot(df['구분배기량'])
    plt.xlabel('Index')
    plt.ylabel('Classified Displacement')
    plt.title('Classified Displacement Graph')

    # 최대값, 최소값 표시
    max_value = df['구분배기량'].max()
    min_value = df['구분배기량'].min()

    plt.annotate(f'Max: {max_value}', xy=(df['구분배기량'].idxmax(), max_value), xytext=(df['구분배기량'].idxmax(), max_value + 10), arrowprops=dict(arrowstyle='->'))
    plt.annotate(f'Min: {min_value}', xy=(df['구분배기량'].idxmin(), min_value), xytext=(df['구분배기량'].idxmin(), min_value - 10), arrowprops=dict(arrowstyle='->'))

    plt.show()
else:
    print("'구분배기량' 컬럼이 숫자형 데이터가 아닙니다. 그래프를 생성할 수 없습니다.")


```
