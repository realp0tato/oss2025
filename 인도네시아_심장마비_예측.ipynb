#나이
조사한 인도네시아 인구의 나이별 분류 그래프


```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 age를 막대그래프로 표시하는데 x축에 age를 표시하는데 단위는 1단위지만 x축에 표시하는 age는 최대랑 최소만 표시해줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=range(int(df['age'].min()), int(df['age'].max()) + 2, 1), kde=False) # bin size = 1
plt.xticks([int(df['age'].min()), int(df['age'].max())]) # Only show min and max age on x-axis
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

#콜레스테롤 수치
조사한 인도네시아인의 콜레스테롤 수치

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 cholesterol_level 을 막대그래프로 보여줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(10, 6))
sns.histplot(df['cholesterol_level'], kde=False)
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.title('Distribution of Cholesterol Level')
plt.show()
```

#혈압
수축기 혈압과 이완기 혈압의 그래프

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 blood_pressure_systolic 과 blood_pressure_diastolic을 사용해서 두개의 막대그래프로 보여줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['blood_pressure_systolic'], kde=False)
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Frequency')
plt.title('Distribution of Systolic Blood Pressure')

plt.subplot(1, 2, 2)
sns.histplot(df['blood_pressure_diastolic'], kde=False)
plt.xlabel('Diastolic Blood Pressure')
plt.ylabel('Frequency')
plt.title('Distribution of Diastolic Blood Pressure')

plt.tight_layout()
plt.show()
```

#가족력
가족력의 그래프

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 family_history를 막대그래프로 보여줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(8, 6))
sns.countplot(x='family_history', data=df)
plt.xlabel('Family History')
plt.ylabel('Frequency')
plt.title('Distribution of Family History')
plt.show()
```

#심장마비


##나이와의 연관성

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 age와 heart_attack을 연관지어서 그래프로 보여줘 age의 x축값은 최대와 최소만표시하고 heart_attack은 1값만 표시해줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Filter for heart attack cases (heart_attack == 1)
heart_attack_df = df[df['heart_attack'] == 1]

# Create the plot.
plt.figure(figsize=(10, 6))
sns.histplot(heart_attack_df['age'], bins=range(int(df['age'].min()), int(df['age'].max()) + 2, 1), kde=False)
plt.xticks([int(df['age'].min()), int(df['age'].max())])  # Only show min and max age on x-axis
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age for Heart Attack Cases')
plt.show()
```

##콜레스테롤 수치와의 연관성

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 cholesterol_level 이랑 heart_attack을 연관지어서 막대그래프로 나타내줘 근데 heart_attack은 1의 값만 표시하고 cholesterol_level의 x축 값은 최대값 최소값만 표시해줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Filter for heart attack cases (heart_attack == 1)
heart_attack_df = df[df['heart_attack'] == 1]

# Create the plot.
plt.figure(figsize=(10, 6))
sns.barplot(x='cholesterol_level', y='heart_attack', data=heart_attack_df, estimator=sum)
plt.xticks([df['cholesterol_level'].min(), df['cholesterol_level'].max()])
plt.xlabel('Cholesterol Level')
plt.ylabel('Number of Heart Attacks')
plt.title('Heart Attacks and Cholesterol Level')
plt.show()
```

##혈압과 심장마비의 연관성
###수축기 혈압

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 blood_pressure_systolic 과 heart_attack을 연관지어서 막대그래프로 표시해줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(10, 6))
sns.barplot(x='blood_pressure_systolic', y='heart_attack', data=df, estimator=lambda x: len(x))
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Heart Attack Count')
plt.title('Systolic Blood Pressure and Heart Attack')
plt.show()
```

##이완기 혈압

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 blood_pressure_diastolic 과 heart_attack을 연관지어서 막대그래프로 보여줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(10, 6))
sns.barplot(x='blood_pressure_diastolic', y='heart_attack', data=df, estimator=lambda x: len(x))
plt.xlabel('Diastolic Blood Pressure')
plt.ylabel('Heart Attack Count')
plt.title('Diastolic Blood Pressure and Heart Attack')
plt.show()
```

##가족력과의 연관성

```python
# prompt: /content/heart_attack_prediction_indonesia.csv 에서 family_history와 heart_attack을 연관지어서 막대그래프로 보여줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv('/content/heart_attack_prediction_indonesia.csv')

# Create the plot.
plt.figure(figsize=(8, 6))
sns.countplot(x='family_history', hue='heart_attack', data=df)
plt.xlabel('Family History')
plt.ylabel('Frequency')
plt.title('Family History and Heart Attack')
plt.show()
```
